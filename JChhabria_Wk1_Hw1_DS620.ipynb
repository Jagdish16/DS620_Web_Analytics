```python
# Import libraries
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create the graph object
G = nx.krackhardt_kite_graph()
# Create the labels object for naming the nodes
labels = {0:'Andre',1:'Beverly',2:'Carol',3:'Diane',4:'Ed',5:'Fernando',6:'Garth',7:'Heather',8:'Ike', 9:'Jane'} 
# Draw the network graph
nx.draw_networkx(G,with_labels=True,labels=labels,font_weight='bold')
# Display the graph
plt.show()
```

```python
print("Nodes:","\n",G.nodes(),"\n") # Lists all the nodes
print("Edges:","\n",G.edges(),"\n")  # Lists all the edges
print("Number of Nodes:","\n",G.number_of_nodes(),"\n") # Shows number of nodes
print("Number of Edges:","\n",G.number_of_edges(),"\n")  # Shows number of edges
```
